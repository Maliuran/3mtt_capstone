```python
def Life_Stage(x):

    if x >=60:
        return 'Senior Adult'
    elif x >=40:
         return 'Mid_Adult'
    elif x >=20:
        return 'Adult'
    elif x >=13:
        return 'Teen'
    elif x >=7:
        return 'Child'
```

<h1 style='text-align:center;'>Eurocontrol</h1>

The four category of flight distances according to Eurocontrtol
1. Very-short-haul flight (<500km)
2. Short-haul flight (500km-1500km)
3. Medium-haul flight (1501km-4000km)
4. Long-haul flight (>4000)

<h1 style='text-align:center;'>A function to categorise the Flight distance</h1>

```python
CS.nsmallest(5,columns='Flight Distance')
```

```python
CS.nlargest(5,columns='Flight Distance')
```

```python
def Distance_Category(y):

    if y >=4000:
         return 'Long_Haul Flight'
    elif y >=1500:
        return 'Medium_Haul Flight'
    elif y >=500:
        return 'Short_Haul Flight'
    elif y >=50:
        return 'Very_Short_Haul Flight'
```

```python
from sklearn.model_selection import train_test_split

X=CS2.drop(columns='satisfaction',)
y=CS2['satisfaction']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y, random_state=8)
```

```python
from sklearn.pipeline import Pipeline
from category_encoders.target_encoder import TargetEncoder
from xgboost import XGBClassifier
from xgboost import plot_importance

estimators= [
    ('encoder', TargetEncoder()),
    ('clf',XGBClassifier(random_state=8))
]
pipe=Pipeline(steps=estimators)
pipe
```

```python
CS2_dummies= pd.get_dummies(CS2,columns=['satisfaction', 'Customer Type', 'Type of Travel','Class'])
```

```python
y=CS2_dummies['satisfaction_satisfied']
X=CS2_dummies.drop(['satisfaction_satisfied','satisfaction_dissatisfied'], axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
xgb=XGBClassifier(objective='binary:logistic', random_state=0)
```

```python
cv_params={'max_dept':[4,6],
          'min_child-weight':[3,5],
          'learning_rate':[0.1,0.2,0.3],
          'n_estimators':[5,10,15],
          'subsample':[0.7],
          'colsample_bytree':[0.9]}
```

```python
scoring = {'accuracy','precision','recall','f1'}
```

```python
xgb_cv=GridSearchCV(xgb,cv_params, cv=5,refit='f1')
```

```python
xgb_cv=xgb_cv.fit(X_train,y_train)
xgb_cv
```

```python
y_pred=xgb_cv.predict(X_test)
y_pred
```

```python
# 1. Print the accuracy score.

ac_score = metrics.accuracy_score(y_test, y_pred)
print('accuracy score:', ac_score)

# 2. Print the precision score.

pc_score = metrics.precision_score(y_test, y_pred)
print('precision score:', pc_score)

# 3. Print the recall score.

rc_score = metrics.recall_score(y_test, y_pred)
print('recall score:', rc_score)


# 4. Print the f1 score.

f1_score = metrics.f1_score(y_test, y_pred)
print('f1 score:', f1_score)
```

```python
# Construct and display the confusion matrix.

# Construct the confusion matrix for the predicted and test values.

cm = metrics.confusion_matrix(y_test, y_pred)

# Create the display for the confusion matrix.

disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=xgb_cv.classes_)

# Plot the visual in-line.

disp.plot()
```

```python
# Plot the relative feature importance of the predictor variables in the model.

plot_importance(xgb_cv.best_estimator_);
```
